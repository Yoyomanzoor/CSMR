```python
from io import StringIO
import numpy as np
import pandas as pd
```

```python
f = open("GSE184303_featureCounts.txt", 'r').read()
f = f.split("\n", 1)[1]
```

```python
df = pd.read_csv(StringIO(f), sep="\t", header=0)
```

```python
df.head()
```

```python
df = df.drop(df.columns[1:6], axis=1)
df.head()
```

```python
# Make a list that will take the names from the file path name. This is done by splitting files by '/' then selecting the 6th element, which is 'KF_(some integer)'.
# Example file path: /project2/ebchang/Katya/ML_analysis/STAR_alignment/KF_46/KF_46_Aligned.sortedByCoord.out.bam
col_names = ["Geneid"]    # define list, already containing the first column so the dimensions will be the same as the dataframe
for i in list(df):
    if i == "Geneid":    # I don't want the column that contains Gene ids
        next
    else:
        label = i.split("/")[6].replace("S","")
        col_names.append(label)

df.columns = col_names

df.head()
```

```python
# Making labels to reorder the dataframe to be, like, actually sensible
new_order = ["Geneid"]
for i in range(0,(len(df.columns)-1)):
    k = i+1
    new_order.append(label.split("_")[0] + f"_{k:02d}")

df = df.reindex(columns=new_order)
df.head()
```
